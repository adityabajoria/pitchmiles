```python
import pandas as pd
serie_a_data = pd.read_csv('/Users/aditya/datascience-projects/soccer-analysis/Brasileirao_Matches.csv')
```

```python
serie_a_data
```

```python
def standardize_team_name(row):
    name = row['home_team']
    state = row['home_team_state']

    # Handle ambiguous names explicitly
    if name.startswith('Atletico'):
        if state == 'MG':
            return 'Atlético Mineiro'
        elif state == 'PR':
            return 'Athletico Paranaense'

    # Fallback: remove the suffix
    return name.replace(f'-{state}', '')

# Apply to both home and away
serie_a_data['home_team'] = serie_a_data.apply(standardize_team_name, axis=1)

def standardize_away_team(row):
    name = row['away_team']
    state = row['away_team_state']

    if name.startswith('Atletico'):
        if state == 'MG':
            return 'Atlético Mineiro'
        elif state == 'PR':
            return 'Athletico Paranaense'

    return name.replace(f'-{state}', '')

serie_a_data['away_team'] = serie_a_data.apply(standardize_away_team, axis=1)
```

```python
serie_a_data
```

```python
serie_a_teams = ['Flamengo', 'Santos', 'Sao Paulo',
                'Cruzeiro', 'Corinthians', 'Palmeiras', 'Botafogo',
                'Gremio', 'Atletico', 'Internacional']

refined_serie_a_data = serie_a_data[
    (serie_a_data['home_team'].isin(serie_a_teams)) &
    (serie_a_data['away_team'].isin(serie_a_teams))
]
```

```python
# drop columns
refined_serie_a_data.drop(columns=['home_team_state', 'away_team_state', 'round'], inplace=True)
```

```python
# move league to be the first column
refined_serie_a_data['league'] = 'Brazilian Serie A'
```

```python
refined_serie_a_data
```

```python
# move league to be the first column
refined_serie_a_data = refined_serie_a_data[['league'] + [col for col in refined_serie_a_data.columns if col != 'league']]
```

```python
refined_serie_a_data = refined_serie_a_data.rename(columns={
    'home_goal': 'home_score',
    'away_goal': 'away_score'
})
```

```python
refined_serie_a_data['datetime'] = pd.to_datetime(refined_serie_a_data['datetime']).dt.date

refined_serie_a_data = refined_serie_a_data.rename(columns={
    'datetime': 'date'
})
```

```python
# drop missing values
refined_serie_a_data.dropna(inplace=True)
```
